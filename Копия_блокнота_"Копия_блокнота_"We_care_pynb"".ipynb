```python
!pip install spacy
!python -m spacy download xx_ent_wiki_sm
import spacy

# NER modelini yuklash
nlp = spacy.load("xx_ent_wiki_sm")

def extract_gender(text):
    text = text.lower()
    if "ayol" in text:
        return "Ayol"
    elif "erkak" in text:
        return "Erkak"
    else:
        return None

# Foydalanish
query = "Andijon viloyatidan, tajribali ayol enaga kerak"
gender = extract_gender(query)
print(gender)  # Ayol
```

```python
def extract_region(text):
    regions = {
        "Andijon": 6, "Farg‘ona": 5, "Namangan": 7, "Toshkent": 1,
        "Samarqand": 2, "Buxoro": 3, "Xorazm": 4, "Qashqadaryo": 8,
        "Surxondaryo": 9, "Jizzax": 10, "Sirdaryo": 11, "Navoiy": 12, "Qoraqalpog‘iston": 13
    }
    for region in regions.keys():
        if region.lower() in text.lower():
            return region, regions[region]
    return None, None

# Foydalanish
region, region_index = extract_region(query)
print(region, region_index)  # Andijon 6
```

```python
def get_nearby_regions(region_index):
    nearby = {6: [5, 7, 6], 5: [4, 6, 7], 7: [6, 8]}  # Misol uchun
    return nearby.get(region_index, [region_index])

# Foydalanish
nearby_regions = get_nearby_regions(region_index)
print(nearby_regions)  # Andijon uchun: [5, 6, 7]
```

```python
def classify_experience(text):
    import re
    match = re.search(r'(\d+)', text)
    if match:
        years = int(match.group(1))
        if years <= 3:
            return "0-3 yil"
        elif years <= 10:
            return "3-10 yil"
        else:
            return "10+ yil"
    return None

# Foydalanish
experience = classify_experience("Andijonda 5 yil tajribali ayol enaga kerak")
print(experience)  # 3-10 yil
```

```python
def search_nanny(query):
    gender = extract_gender(query)
    region, region_index = extract_region(query)
    experience = classify_experience(query)

    # 1. Faqat mos jinsni tanlash
    results = df[df["JINS"] == gender] if gender else df

    # 2. Mos viloyatlarni tanlash (yaqin hududlar qo‘shiladi)
    nearby_regions = get_nearby_regions(region_index)

    # Accessing the 'regions' dictionary from within the 'extract_region' function
    regions = { # This line is added to get the regions dictionary
        "Andijon": 6, "Farg‘ona": 5, "Namangan": 7, "Toshkent": 1,
        "Samarqand": 2, "Buxoro": 3, "Xorazm": 4, "Qashqadaryo": 8,
        "Surxondaryo": 9, "Jizzax": 10, "Sirdaryo": 11, "Navoiy": 12, "Qoraqalpog‘iston": 13
    }
    results = results[results["VILOYAT"].isin([region] + [key for key, val in regions.items() if val in nearby_regions])] if region else results # Changed 'extract_region.items()' to 'regions.items()'

    # 3. Tajribaga qarab filter qilish
    if experience:
        if experience == "0-3 yil":
            results = results[results["TAJRIBA"] <= 3]
        elif experience == "3-10 yil":
            results = results[(results["TAJRIBA"] > 3) & (results["TAJRIBA"] <= 10)]
        elif experience == "10+ yil":
            results = results[results["TAJRIBA"] > 10]
```

```python
def search_nanny(query):
    gender = extract_gender(query)
    region, region_index = extract_region(query)
    experience = classify_experience(query)

    # 1. Faqat mos jinsni tanlash
    results = df[df["JINS"] == gender] if gender else df

    # 2. Mos viloyatlarni tanlash (yaqin hududlar qo‘shiladi)
    nearby_regions = get_nearby_regions(region_index)

    # Accessing the 'regions' dictionary from within the 'extract_region' function
    regions = { # This line is added to get the regions dictionary
        "Andijon": 6, "Farg‘ona": 5, "Namangan": 7, "Toshkent": 1,
        "Samarqand": 2, "Buxoro": 3, "Xorazm": 4, "Qashqadaryo": 8,
        "Surxondaryo": 9, "Jizzax": 10, "Sirdaryo": 11, "Navoiy": 12, "Qoraqalpog‘iston": 13
    }
    results = results[results["VILOYAT"].isin([region] + [key for key, val in regions.items() if val in nearby_regions])] if region else results # Changed 'extract_region.items()' to 'regions.items()'

    # 3. Tajribaga qarab filter qilish
    if experience:
        if experience == "0-3 yil":
            results = results[results["TAJRIBA"] <= 3]
        elif experience == "3-10 yil":
            results = results[(results["TAJRIBA"] > 3) & (results["TAJRIBA"] <= 10)]
        elif experience == "10+ yil":
            results = results[results["TAJRIBA"] > 10]

    return results[["FIO", "VILOYAT", "JINS", "TAJRIBA"]].head(10)
```
